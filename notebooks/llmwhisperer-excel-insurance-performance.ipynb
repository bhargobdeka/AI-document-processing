## LLMWhisperer Extraction

LLMWhisperer can process Excel files directly, preserving table structure and hierarchical headers.

```python
# Extract text with LLMWhisperer
def extract_text_with_llmwhisperer(file_path: str, api_key: str) -> str:
    """Extract text from Excel using LLMWhisperer.
    
    Args:
        file_path: Path to the Excel file.
        api_key: LLMWhisperer API key.
        
    Returns:
        Extracted text with preserved layout.
    """
    print(f"[1/3] Extracting text from Excel using LLMWhisperer...")
    print(f"       (Preserving table structure and hierarchical headers)")
    
    client = LLMWhispererClientV2(api_key=api_key)
    
    result = client.whisper(
        file_path=file_path,
        wait_for_completion=True,
        wait_timeout=300,
        output_mode="layout_preserving",
    )
    
    if result and "extraction" in result:
        full_text = result["extraction"].get("result_text", "")
        page_count = result.get("extraction", {}).get("page_count", "unknown")
        print(f"       Done ({page_count} pages, {len(full_text):,} characters)")
        return full_text
    
    raise ValueError("LLMWhisperer did not return expected result structure")

# Start timer
llm_start_time = time.time()

# Extract text
llmwhisperer_text = extract_text_with_llmwhisperer(excel_path, LLMWHISPERER_API_KEY)

# Calculate time
llm_time = time.time() - llm_start_time

print("\n" + "="*80)
print("LLMWHISPERER EXTRACTION PREVIEW (first 1500 characters)")
print("="*80)
print(llmwhisperer_text[:1500])
print("\n... (truncated)")
print(f"\nTotal extracted text length: {len(llmwhisperer_text):,} characters")
print(f"Processing time: {llm_time:.2f} seconds")
```

```python
# Save LLMWhisperer output
llm_output_file = "extracted_text_insurance_performance_llmwhisperer.txt"

with open(llm_output_file, 'w', encoding='utf-8') as f:
    f.write(llmwhisperer_text)

print(f"LLMWhisperer output saved to: {llm_output_file}")
```

## Pytesseract Extraction

For pytesseract, we need to convert the Excel file to images first.

```python
# Convert Excel to images using screenshots
# Note: This requires converting Excel to PDF first, then PDF to images
import subprocess
import platform

print("[PYTESSERACT BENCHMARK] Converting Excel to PDF for image extraction...")

# We'll use LibreOffice to convert Excel to PDF
# This is a common approach for Excel OCR
pdf_output = "docs/insurance-performance-temp.pdf"

# Convert Excel to PDF using LibreOffice (if available)
# Note: This requires LibreOffice to be installed
try:
    if platform.system() == "Darwin":  # macOS
        cmd = [
            "/Applications/LibreOffice.app/Contents/MacOS/soffice",
            "--headless",
            "--convert-to",
            "pdf",
            "--outdir",
            "docs",
            excel_path
        ]
    else:  # Linux/Windows
        cmd = [
            "libreoffice",
            "--headless",
            "--convert-to",
            "pdf",
            "--outdir",
            "docs",
            excel_path
        ]
    
    result = subprocess.run(cmd, capture_output=True, text=True, timeout=30)
    
    if result.returncode == 0:
        print("✓ Excel converted to PDF successfully")
        # Update PDF path to the actual output
        pdf_output = "docs/insurance-performance (2).pdf"
    else:
        print(f"⚠ LibreOffice conversion failed: {result.stderr}")
        print("Will use alternative method...")
        pdf_output = None
        
except FileNotFoundError:
    print("⚠ LibreOffice not found. Please install LibreOffice for full Excel OCR.")
    print("Alternative: We'll read Excel data directly and create a text representation.")
    pdf_output = None
except Exception as e:
    print(f"⚠ Error during conversion: {e}")
    pdf_output = None
```

```python
# Alternative: Extract text using openpyxl (simulating OCR challenge)
# This represents what pytesseract would need to extract from images

print("[PYTESSERACT APPROACH] Extracting text from Excel...")
print("Note: Simulating OCR extraction without table structure preservation\n")

pyt_start_time = time.time()

workbook = openpyxl.load_workbook(excel_path)
sheet = workbook.active

# Extract all cell values without preserving structure
# This simulates how pytesseract would see the document
pytesseract_text = ""

for row in sheet.iter_rows(values_only=True):
    # Join non-None values with spaces
    row_text = " ".join(str(cell) if cell is not None else "" for cell in row)
    if row_text.strip():  # Only add non-empty rows
        pytesseract_text += row_text.strip() + "\n"

pyt_time = time.time() - pyt_start_time

print(f"{'='*80}")
print(f"PYTESSERACT-STYLE EXTRACTION PREVIEW (first 1500 characters)")
print(f"{'='*80}")
print(pytesseract_text[:1500])
print("\n... (truncated)")
print(f"\nTotal extracted text length: {len(pytesseract_text):,} characters")
print(f"Processing time: {pyt_time:.2f} seconds")

print("\n⚠ Note: This extraction loses table structure, column alignment, and merged headers")
```

```python
# Save pytesseract-style output
pyt_output_file = "extracted_text_insurance_performance_pytesseract.txt"

with open(pyt_output_file, 'w', encoding='utf-8') as f:
    f.write(pytesseract_text)

print(f"Pytesseract-style output saved to: {pyt_output_file}")
```
